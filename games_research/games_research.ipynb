 Также необходимо отформатировать названия столбцов.

Отформатируем названия столбцов.

```python
new_names = {} #словарь новых названий

for name in data.columns:
    new_names[name] = name.lower()#для новых названий приводим старые к нижнему регистру

data = data.rename(columns=new_names)
data.head(2)
```

#### Преобразование типов

Так как год выпуска является натуральным числом, приведем данные в столбце `year_of_release` к типу *int*:

```python
data['year_of_release'] = data['year_of_release'].astype('int', errors='ignore')
```

Оценка критиков является целым числом в диапазоне 0-100, поэтому тип данных для `critic_score` так же *int*:

```python
data['critic_score'] = data['critic_score'].astype('int', errors='ignore')
```

Оценка пользователей является числом с 1 десятичным знаком, поэтому тип данных для use_score - *float*:

```python
data['user_score'] = data['user_score'].astype('float', errors='ignore')
```

```python
data.info()
```

#### Обработка пропусков

В столбце с названиями игр всего 2 пропуска:

```python
data[data['name'].isna()]
```

В данных строках мало информации, избавимся от них:

```python
data = data.dropna(subset=['name'])
```

В столбце с данными о годе выпуска есть пропуски:

```python
data['year_of_release'].isna().sum()
```

```python
data[data['year_of_release'].isna()].head(5)
```

Многие игры выходят одновременно на нескольких платформах. Заполним при возможности год выпуска для мультиплатформенных игр:

```python
no_year_data = data[data['year_of_release'].isna()]
names = list(data.loc[data['year_of_release'].isna(), 'name'])
for name in names:
    data.loc[(data['name'] == name), 'year_of_release'] =\
    data.loc[data['name'] == name, 'year_of_release'].fillna(data.loc[data['name'] == name, 'year_of_release'].max())       
```

После обработки остались пропуски:

```python
data['year_of_release'].isna().sum()
```

Некоторые игры в названии содержат год, например, *Madden NFL 2004*. Заполним пропуски значением из названия:

```python
def get_year(name):
    year = float('nan')
    last_word = name.split()[-1]
    if last_word.isdigit():
        y = int(last_word)        
        if y >= 1980 and y <= 2016: #
            year = y
        if y >= 80 and y < 100: #игры до 2000
            year = y + 1900
        if y >= 5 and y <= 16: #игры после 2000г без тех, для которых число означает номер части 
            year = y + 2000
    
    return year

data.loc[data['year_of_release'].isna(), 'year_of_release'] = data.loc[data['year_of_release'].isna(), 'name'].apply(get_year)
```

```python
data['year_of_release'].isna().sum()
```

Всего 16713 строк в данных. 138 составляют менее 1%. Избавимся от них: 

```python
data = data.dropna(subset=['year_of_release'])
```

В столбцах с оценками пропущено около половины значений. Оценка пользователей может иметь значение `tbd` - 'подлежит уточнению'. Строковые значения в столбце с числами могут помешать исследованию, заменим их на NaN.

```python
data['user_score'] = data['user_score'].replace('tbd', np.nan)
data['user_score'] = data['user_score'].astype('float', errors='ignore')
```

Оценки пользователей и критиков можно попробовать восстановить. Исследуем зависимость между оценками критиков и пользователей: 

```python
data['user_score'].corr(data['critic_score']).round(2)
```

Корреляция средняя. Проверим также зависимость для самых популярных платформ: 

```python
rows = {}
for platform in data['platform'].unique():
    if len(data[data['platform'] == platform]) > 500:
        sample = data[data['platform'] == platform]
        rows[platform] = sample['user_score'].corr(sample['critic_score'])
        
pd.DataFrame(rows.values(), index=rows.keys(), columns=['r'])
```

Корреляция для отдельных платформ также является средней. Проверим корреляцию для отдельных жанров:

```python
rows = {}
for genre in data['genre'].unique():
    if len(data[data['genre'] == genre]) > 500:
        sample = data[data['genre'] == genre]
        rows[genre] = sample['user_score'].corr(sample['critic_score'])
pd.DataFrame(rows.values(), index=rows.keys(), columns=['r'])
```

Корреляция средняя. Проверим корреляцию между оценками для отдельных лет:

```python
rows={}
for year in range(2010, 2017):
    sample = data[data['year_of_release'] == year]
    rows[year] = sample['user_score'].corr(sample['critic_score'])
    
pd.DataFrame(rows.values(), index=rows.keys(), columns=['corr'])
```

Корреляция средняя. Проверим корреляцию между оценками для разных рейтингов:

```python
rows = {}
for rating in data['rating'].unique():
    if len(data[data['rating'] == rating]) > 100:
        sample = data[data['rating'] == rating]
        rows[rating] = sample['user_score'].corr(sample['critic_score'])
        
pd.DataFrame(rows.values(), index=rows.keys(), columns=['r'])
```

Корреляция все еще средняя. Введем дополнительный столбец с суммарными продажами и попробуем оценить зависимость между оценками и продажами:

```python
columns = ['na_sales', 'eu_sales', 'jp_sales', 'other_sales']
data['total_sales'] = data[columns].sum(axis=1)
data['user_score'].corr(data['total_sales']).round(2)
```

```python
data['critic_score'].corr(data['total_sales']).round(2)
```

Для оценок пользователей корреляция является очень слабой, для оценок критиков - слабой. Попробуем предсказать оценки на основе продаж отдельно для Северной Америки, Европы, Японии и остальных. При исследовании будем использовать только ненулевые продажи, так как, скорее всего, в таком случае игра просто не была в продаже в регионе:

```python
reg = ['Америка','Европа','Япония','Другие']
reg_sales = ['na_sales', 'eu_sales', 'jp_sales', 'other_sales']
rows = {}

for i in range(4):
    sample = data[data[reg_sales[i]] > 0]
    rows[reg[i]] = sample['critic_score'].corr(sample[reg_sales[i]])

print('Корреляция оценок критиков и продаж по регионам:')    
pd.DataFrame(rows.values(), index=rows.keys(), columns=[''])
```

```python
rows = {}
for i in range(4):
    sample = data[data[reg_sales[i]] > 0]
    rows[reg[i]] = sample['user_score'].corr(sample[reg_sales[i]])
    
print('Корреляция оценок пользователей и продаж по регионам:')
pd.DataFrame(rows.values(), index=rows.keys(), columns=['corr'])
```

Корреляция все еще является слабой. Таким образом ни для оценок пользователей, ни для оценок критиков ни по одному из исследуемых признаков не получено высокой или очень высокой корреляции. Поэтому определять оценки по одному из этих критериев для заполнения пропусков данных не стоит. Оставляем пропуски незаполненными.

Разберем пропуски в данных о рейтинге. Всего пропусков:

```python
data['rating'].isna().sum()
```

Рейтинг ESRB появился в 1994 году. Играм, вышедших ранее, данный рейтинг не мог быть присвоен. Для таких игр заменим пропуски в рейтинге на `N/A`:

```python
data.loc[data['year_of_release'] <= 1994, 'rating'] = data.loc[data['year_of_release'] <= 1994, 'rating'].fillna('N/A')
```

```python
data['rating'].isna().sum()
```

Рейтинг `RP` присваивается, когда когда рейтинг еще неизвестен. Заполним оставшиеся пропуски этим значением:

```python
data['rating'] = data['rating'].fillna('RP')
```

Исследуем, как влияют на продажи внутри одной популярной платформы отзывы пользователей и критиков. Построим диаграммы рассеивания и посчитаем коэффициенты корреляции между отзывами и продажами.

##### PS4

```python
def research(data):
    level_90 = data['total_sales'].quantile(q=.90)
    user_score_mean = data['user_score'].mean()
    critic_score_mean = data['critic_score'].mean()

    print('Количество игр:', len(data))
    print()
    print('Средняя оценка пользователей:', round(user_score_mean, 1))
    print('Средняя оценка критиков:', round(critic_score_mean, 0))
    print()
    print('0.9-квантиль:', level_90.round(2), 'млн.шт')
    
    fig, ax = plt.subplots(nrows=1, ncols=2)
    fig.set_figheight(6)
    fig.set_figwidth(15)

    data.plot(x='critic_score', y='total_sales', style='.', ax=ax[0])
    pd.pivot_table(data, index='critic_score', values='total_sales', aggfunc='mean').plot(grid=True, ax=ax[0])
    ax[0].axhline(y=level_90, color='r', linestyle='--', lw=1) #add average sales
    ax[0].axvline(x=critic_score_mean, color='r', linestyle='--', lw=1) #add average score

    data.plot(x='user_score', y='total_sales', style='.', ax=ax[1])
    pd.pivot_table(data, index='user_score', values='total_sales', aggfunc='mean').plot(grid=True, ax=ax[1])
    ax[1].axhline(y=level_90, color='r', linestyle='--', lw=1) #add average sales
    ax[1].axvline(x=user_score_mean, color='r', linestyle='--', lw=1) #add average score
    plt.show()
    
    # Подсчет коэффициентов корреляции
    print('Коэффициент корреляции общих продаж и оценки критиков:',\
          round(data['total_sales'].corr(data['critic_score']), 2))
    print('Коэффициент корреляции общих продаж и оценки пользователей:',\
          round(data['total_sales'].corr(data['user_score']), 2))
    print('Коэффициент корреляции оценок:',\
          round(data['critic_score'].corr(data['user_score']), 2))


data_ps4 = actual_data.query('platform == "PS4"')

research(data_ps4)
```

Корреляция между оценками критиков и продажами слабая, а между оценками пользователей и продажами практически отутствует. 
На графиках отмечен уровень продаж 2.14 млн.шт, который не превышают 90% игр. Ниже этого уровня для игр отсутствует явный тренд изменение продаж в зависимости от оценок. Топ-10 игр с наиболее высокими продажами имеют высокие оценки критиков:

```python
data_ps4.dropna(subset=['critic_score']).sort_values(by='total_sales', ascending=False).head(10)
```

**Вывод:** Для платформы PS4 оценка критиков оказалась более надежным показателем будущих продаж, и самые популярные игры имеют довольно высокие оценки критиков(выше медианы). Однако более высокая оценка не гарантирует больший объем продаж. Также важно отметить, что в топе почти все игры являются продолжениями/дополнениям/обновлениями существующих игр, и, вероятно, отчасти их популярность связана с известностью предыдущих частей.

```python
# групировка по жанрам и платформам, сортировка
genres_data = actual_data.groupby(['genre', 'platform'], as_index=False)\
.agg({'total_sales': 'sum'})\
.sort_values('total_sales', ascending=False)
```

Чтобы оценить долю игр, приходящую на каждый жанр, необходимо для каждой платформы определить суммарное количество продаж и поделить продажи на это значение.

```python
genres_data['rate'] = 0
for platform in genres_data['platform'].unique():
    total = genres_data.loc[genres_data['platform'] == platform, 'total_sales'].sum()
    genres_data.loc[genres_data['platform'] == platform, 'rate'] =\
    genres_data.loc[genres_data['platform'] == platform, 'total_sales'] / total
```

### Проверка гипотез

#### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

>**Нулевая гипотеза:** средние пользовательские рейтинги платформ Xbox One и PC равны.
>
>**Альтернативная гипотеза:** средние пользовательские рейтинги платформ Xbox One и PC не равны.

```python
#выбираем данные за актуальный период - с 2012 года
data_xone = data.query('platform == "XOne"').dropna(subset=['user_score'])
data_pc = data.query('platform == "PC" and year_of_release >= 2012').dropna(subset=['user_score'])

alpha = 0.05 #критический уровень статистической значимости

results = st.ttest_ind(
    data_xone['user_score'], 
    data_pc['user_score'],
    equal_var = False) #проверьте гипотезу о равенстве средних двух независимых совокупностей

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

```python
print('Средняя оценка для XOne', round(data_xone['user_score'].mean(), 2))
print('Средняя оценка для PC', round(data_pc['user_score'].mean(), 2))
```

Средние оценки в выборках отличаются менее чем на 0.1.

#### Средние пользовательские рейтинги жанров Action и Sports разные

>Нулевая гипотеза: средние пользовательские рейтинги жанров Action и Sports равны.
>
>Альтернативная гипотеза: средние пользовательские рейтинги жанров Action и Sports не равны.

```python
data_action = data.query('genre == "Action" and year_of_release >= 2012').dropna(subset=['user_score'])
data_sports = data.query('genre == "Sports" and year_of_release >= 2012').dropna(subset=['user_score'])
```

```python
alpha = 0.05 #критический уровень статистической значимости

results = st.ttest_ind(
    data_action['user_score'], 
    data_sports['user_score'],
    equal_var = False) # проверка гипотезы о равенстве средних двух независимых совокупностей

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```
